```python
import anndata
adata_ref = anndata.read_h5ad('/nfs/team205/ds40/G2G/MS/Datasets/RefATOAnalysis/PanfetalATO_SPT_gplvm_ref_03012023.h5ad')
adata_query = anndata.read_h5ad('/nfs/team205/ds40/G2G/MS/Datasets/RefATOAnalysis/PanfetalATO_SPT_gplvm_ATO_03012023.h5ad')
adata_ref.obs['ID'] = np.repeat('Reference',adata_ref.shape[0])
adata_query.obs['ID'] = np.repeat('Organoid',adata_query.shape[0])
adata_all = anndata.concat([adata_ref, adata_query])
```

# Plot pseudotime stripplot

```python
ref_cmap = {
        'CYCLING_MPP': vega_20[12],
        'HSC_MPP':vega_20[13],
        'LMPP_MLP':vega_20[10],
        'DN(early)_T': vega_20[0],
        'DN(P)_T':vega_20[1],
        'DN(Q)_T':vega_20[2],
        'DP(P)_T': vega_20[3],
        'DP(Q)_T':vega_20[4],
        'ABT(ENTRY)':vega_20[5],
        'TYPE_1_INNATE_T':vega_20[9]
       }
ato_cmap = { 
            'PRIMITIVE_STREAK': vega_20[19],
            'IPSC':  vega_20[15],
            'HSC_MPP':  vega_20[13],
            'HSC_MPP/LMPP_MLP/DC2':  vega_20[10],
            'MESODERMAL_PROGENITOR':  vega_20[17],
            'ENDOTHELIUM':vega_20[18],
            'DN(EARLY)_T':vega_20[0],
            'DN_T':vega_20[1],
            'DP(P)_T':vega_20[3],
            'DP(Q)_T':vega_20[4],
           'ABT(ENTRY)':vega_20[5],
            'SP_T':vega_20[7]
           }
```

```python
adata_ref.obs['ANNOTATION_COMB'] = adata_ref.obs['ANNOTATION_COMB'].cat.reorder_categories(['CYCLING_MPP','HSC_MPP','LMPP_MLP','DN(early)_T',
                                                                                            'DN(P)_T','DN(Q)_T','DP(P)_T','DP(Q)_T','ABT(ENTRY)',
                                                                                            'TYPE_1_INNATE_T'])
```

```python
fig, ax = plt.subplots(figsize=(8,2))
sb.set_theme(style='ticks')
ax=sb.stripplot(data=adata_ref.obs, x="time", y="ANNOTATION_COMB", hue="ANNOTATION_COMB",
                   palette=ref_cmap,jitter=True, size=2)
ax.set_ylabel('')    
ax.set_xlabel('pseudotime', fontsize=10)
ax.set_title('Pan fetal reference')
ax.get_legend().remove()
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
ax.xaxis.grid(False)

plt.savefig(fig_path+'ref_pseudotime_stripplot.pdf',bbox_inches='tight')
```
