```python
# construct the empirical null distribution
np.random.seed(0)
null_dist = []
for g in tqdm(human_TFs):  
        gex_ref = np.asarray(R[:, g].X.todense()).flatten() 
        gex_ato = np.asarray(Q[:, g].X.todense()).flatten() 
        
        for i in range(50):
            x = np.random.choice(gex_ref, 100, replace = False)
            y = np.random.choice(gex_ato, 100, replace = False)

            d = compute_mmldist(x[0:50], x[51:100])
            null_dist.append(d)
            d = compute_mmldist(y[0:50], y[51:100])
            null_dist.append(d) 
```

```python
def compute_mean_expression_diffs(GENELIST, gene_dists_refato, gene_dists_refatotnf):
    # ALL HUMAN TFS --- under GAMMA empirical NULL Dist
    temp_r = []
    temp_ato = []
    temp_tnf = []
    for g in tqdm(GENELIST):
            gex1 = np.asarray(R[:, g].X.todense()).flatten() 
            gex2 = np.asarray(Q[:, g].X.todense()).flatten() 
            gex3 = np.asarray(adata_tnf[:, g].X.todense()).flatten() 
            temp_r.append(np.mean(gex1))
            temp_ato.append(np.mean(gex2) ) 
            temp_tnf.append(np.mean(gex3) ) 

    df = pd.DataFrame([GENELIST, gene_dists_refato, gene_dists_refatotnf, temp_r, temp_ato, temp_tnf]).transpose()
    df.columns = ['TF', 'ref_ato','ref_ato_tnf','mean_ref','mean_ato','mean_tnf']
    df = df.set_index('TF')
    df['diff'] = df['ref_ato'] - df['ref_ato_tnf']
    df = df.sort_values('diff', ascending=False)
    df['rank'] = np.arange( 0, df.shape[0] ) 
    return df

def compute_empirical_cdf_pvalues(gene_dists_refato,gene_dists_refatotnf):
    np.random.seed(0)
    ecdf = statsmodels.distributions.empirical_distribution.ECDF(null_dist)
    # compute p-value given null distribution of distances
    p_values1 = []
    p_values2 = []
    for i in tqdm(range(len(gene_dists_refato))):
        p_values1.append(estimate_pvalue(gene_dists_refato[i], null_dist, a,loc,scale, plot=False, empirical_cdf=True, ecdf=ecdf))
        p_values2.append(estimate_pvalue(gene_dists_refatotnf[i], null_dist, a, loc, scale, plot=False, empirical_cdf=True, ecdf=ecdf))

    temp = pd.DataFrame([p_values1, p_values2]).transpose() 
    temp.index = GENELIST
    temp.columns = ['pval_b4','pval_aftr']
    # p-value correction for FDR
    temp['pval_b4'] = statsmodels.stats.multitest.fdrcorrection(temp['pval_b4'])[1]
    temp['pval_aftr'] = statsmodels.stats.multitest.fdrcorrection(temp['pval_aftr'])[1]
    return temp 

def compute_gamma_cdf_pvalues(gene_dists_refato,gene_dists_refatotnf):
    p_values1 = []
    p_values2 = []
    for i in tqdm(range(len(gene_dists_refato))):
        p_values1.append(estimate_pvalue(gene_dists_refato[i], null_dist, a,loc,scale, plot=False, empirical_cdf=False))
        p_values2.append(estimate_pvalue(gene_dists_refatotnf[i], null_dist, a, loc, scale, plot=False, empirical_cdf=False))

    temp = pd.DataFrame([p_values1, p_values2]).transpose() 
    temp.index = GENELIST
    temp.columns = ['pval_b4','pval_aftr']
    # p-value correction for FDR
    temp['pval_b4'] = statsmodels.stats.multitest.fdrcorrection(temp['pval_b4'])[1]
    temp['pval_aftr'] = statsmodels.stats.multitest.fdrcorrection(temp['pval_aftr'])[1]
    return temp 
    
def print_stat(temp, df, PVAL_THRESHOLD):

    print('genes ', len(temp))
    print('# of siginificant different genes before: ',temp[temp['pval_b4']<PVAL_THRESHOLD].shape[0]) 
    print('# of siginificant different genes after: ',temp[temp['pval_aftr']<PVAL_THRESHOLD].shape[0]) 
    
    
    common_DE = np.intersect1d( temp[temp['pval_aftr']<PVAL_THRESHOLD].index, temp[temp['pval_b4']<PVAL_THRESHOLD].index ) 
    print('# of common genes which were significantly different before and after both: ', len(common_DE)   )
    print(common_DE)
    if(len(common_DE)>0):
        # on the path to getting closer to ref even if they are still different
        test = df.loc[common_DE]
        glist = list(common_DE) 
        z = [] 
        for i in range(test.shape[0]):
            if(test['mean_ato'][i] < test['mean_tnf'][i] and test['mean_tnf'][i] < test['mean_ref'][i]):
                z.append(glist[i])
            if(test['mean_ref'][i] < test['mean_tnf'][i] and test['mean_tnf'][i] < test['mean_ato'][i]):
                z.append(glist[i])
        print('Out of the common DE, ones on the path to getting closer: ', len(z))
        print(z)

    filter_df = temp[np.logical_and(temp['pval_b4'] < PVAL_THRESHOLD  , temp['pval_aftr'] > PVAL_THRESHOLD )]
    print('# of genes significantly different before, but now have become closer to ref with no significant difference: ', filter_df.shape[0] )
    print(filter_df.index)

    filter_df = temp[np.logical_and(temp['pval_b4'] > PVAL_THRESHOLD  , temp['pval_aftr'] < PVAL_THRESHOLD )]
    print('# of genes similar before, but now have become different to ref with significant difference: ', filter_df.shape[0] )
    print(filter_df.index)
    print('Out of these ', filter_df.shape[0]  , ' differences: ')
    glist = list(filter_df.index) 
    df.loc[glist]

    test = df.loc[glist]
    x = []; y = [];
    x1 = []; x2=[]; 
    for i in range(test.shape[0]):

        # CHECK THE POSSIBLE UP AND DOWN groups of genes
        # ATO < REF < TNF  # suggests need of adjustment of concentration
        if(test['mean_ato'][i] < test['mean_ref'][i] and  test['mean_ref'][i] < test['mean_tnf'][i]):
            x.append(glist[i])
            x1.append(glist[i])
        # TNF < REF < ATO  # suggests need of adjustment of concentration
        if(test['mean_tnf'][i] < test['mean_ref'][i] and test['mean_ref'][i] < test['mean_ato'][i]):
            x.append(glist[i])
            x2.append(glist[i])

        # REF < ATO < TNF # increases the expression of current ATO towards opposite to ref
        if(test['mean_ref'][i] < test['mean_ato'][i] and  test['mean_ato'][i] < test['mean_tnf'][i]):
            y.append(glist[i])
        # TNF < ATO < REF # decreases the expression of current ATO towards opposite to ref
        if(test['mean_tnf'][i] < test['mean_ato'][i] and  test['mean_ato'][i] < test['mean_ref'][i]):
            y.append(glist[i])

    print('# of genes that surpass the ref in both directions (suggesting modulation of concentration): ', len(x))
    print(x)
    print('# of genes which became opposite to ref: ', len(y))
    print(y)
    
    print('********************************************************************')
    
    significant_distant_TFs_before = df.loc[temp[temp['pval_b4']<PVAL_THRESHOLD].index] # significantly different TFs before TNFa
    significant_distant_TFs_after = df.loc[temp[temp['pval_aftr']<PVAL_THRESHOLD].index] # significantly different TFs before TNFa

    a = list(significant_distant_TFs_before ['ref_ato'])
    b =  list(significant_distant_TFs_after ['ref_ato_tnf'])
    print('mean distance: ', round(np.mean(a),2),round(np.mean(b),2) )
    print('median distance: ', np.median(a),np.median(b) )
    print(stats.ttest_ind(a,b, alternative='greater'))
    print(stats.mannwhitneyu(a,b, alternative='greater'))
    print(stats.ranksums(a,b, alternative='greater'))
```

```python
#Due to technical effects, we have noise the decreases and increases of the distances we see. 
#To avoid bias in noise, we select significant distances in both cases. 
#If a gene show significantly different decrease only after tnf, its distance taken
#If a gene show significantly different increase only after tnf, its distance taken
```
